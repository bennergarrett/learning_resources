```python
import re

def anagram(s, t):
    s=s.lower().strip().replace(" ", "")
    t=t.lower().strip().replace(" ", "")
    count = 1
    anagrams = []
    an=''
    ##reverses string
    ##
    anagrams.append(s)
    #anagrams.append(s[::-1])
    #anagrams.append(reverse_word(s))
    #print(an)
    #anagrams.append(an)
    while count < (len(s)):
        ##initial switch
        ##
        
        letters = s[-count::-1]
        print("letters are: " + letters)
        
        start = letters + re.sub("|".join(letters), "", s)
        print(count)
        anagrams.append(start)
        #anagrams.append(reverse_word(start))
        print("start is: " + start)
        an = start[-1] + start[0:-1]
        anagrams.append(an)
        #anagrams.append(reverse_word(an))
        print("an is: " + an)
        
        
        while an != start:
            ##initial store of first possible anagram
            ##
            ##switching letters through
            ##
            an = an[-1] + an[0:-1]
            anagrams.append(an)
            #anagrams.append(reverse_word(an))
            print(an)
        #print(anagrams)
        count += 1
    

    
    print(anagrams)
            
        
  
    for possibles in anagrams:
        ##print(possibles)
        if possibles == t:
            return True
    return False


def reverse_word(word):
    new_word = []
    final = ""
    for i in range(len(word)):
        new_word.append(word[-i])
    final = new_word
    return final
    

    
        
        
    
    
```
